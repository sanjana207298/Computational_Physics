### (ùëñ) Find the smallest number that is not zero

```python
# Initializing a variable 'n' with the smallest positive float64 value
n = np.float64(1)

# Keep dividing 'n' by 2 until it becomes '0' (or as close as possible)
while n > 0:
    n = n / 2
    print(n)

# Theoretically, there's no smallest non-zero number, but due to limited memory,
# Python considers numbers smaller than 5e-324 as '0' in float64 representation.
# So, in this case, the smallest non-zero number is approximately 5e-324.
```

### (ùëñùëñùëñ) ùêñùê°ùêöùê≠ ùê¢ùê¨ ùê≠ùê°ùêû ùê•ùêöùê´ùê†ùêûùê¨ùê≠ ùêØùêöùê•ùêÆùêû ùê®ùêü b ùêüùê®ùê´ ùê∞ùê°ùê¢ùêúùê° ùê≠ùê°ùêû ùêÆùê¨ùêÆùêöùê• ùê©ùêöùê¢ùê´ ùê®ùêü ùê¨ùê®ùê•ùêÆùê≠ùê¢ùê®ùêßùê¨, $‚àí ùëè \pm \sqrt{(ùëè^2 + 1)}$ ùêúùêöùêß ùêõùêû ùêÆùê¨ùêûùêù (ùê¨ùêöùê≤, ùêüùê®ùê´ ùêö 32-ùêõùê¢ùê≠ ùê¶ùêöùêúùê°ùê¢ùêßùêû)?

The largest value of $b$ for which the usual pair of solutions $(-b \pm \sqrt{b^2 + 1})$ can be used safely (e.g., for a 32-bit machine) is determined by the limits of floating-point representation. In a typical 32-bit floating-point representation, you have 24 bits of precision for the mantissa (fractional part), 8 bits for the exponent, and 1 bit for the sign (assuming that the computer uses 2's complement).

For the formula $(-b \pm \sqrt{b^2 +1})$, the critical consideration is the evaluation of $b^2 +1 $ under these limitations. The largest value of $b$ that can be used without causing overflow or loss of significant digits is when $b^2 +1$ can still be accurately represented within the available precision.

Here's the calculation to find the largest value of $b$:

Precision Limitation: In a 32-bit floating-point representation, the mantissa has 24 bits of precision. This means that the largest integer that can be exactly represented as a 32-bit float is $2^{24} - 1$.

Determine $b$: The equation $b^2 +1$ should not exceed the maximum representable integer value.

So, you want to find $b$ such that:
$b^2 +1 \leq 2^{24} - 1$

Rearrange the equation: Solve for $b$:
$
b \leq \sqrt{2^{24} - 1 -1}
$

So, the largest value of $b$ that can be safely used in the quadratic formula $(-b \pm \sqrt{b^2 +1})$ for a 32-bit machine with the given precision limits, calculated using the above formula is 16777214.



## Question 3. Conversion

### decimal integer $\rightarrow$ binary

```python
#Only for positive
try:
    num = int(input('Enter a positive integer decimal number: '))
    if num < 0:
        raise ValueError("Please enter a positive integer.")
except ValueError as e:
    print(e)
else:
    binary = ""
    while num > 0:
        binary = str(num % 2) + binary
        num //= 2
    print(f'The binary form of {num} is: {binary}')
```

```python
#Positive and Negative
def decimal_to_binary(num):
    if num < 0:
        # Calculate the two's complement of a negative number
        num = 2**32 + num  # Assuming 32-bit integers

    binary = ""
    while num > 0:
        binary = str(num % 2) + binary
        num //= 2

    return binary

try:
    num = int(input('Enter an integer: '))
    binary = decimal_to_binary(num)
    print(f'The binary form of {num} is: {binary}')
except ValueError as e:
    print(e)
```

### decimal floating point number $\rightarrow$ binary

```python
#Only for positive
try:
    decimal = float(input('Enter a decimal number: '))
except ValueError as e:
    print('Invalid input:', e)
else:
    if decimal < 0:
        print("Please enter a non-negative decimal number.")
    else:
        integer_part = int(decimal)
        fractional_part = decimal - integer_part

        binary_integer = ""
        while integer_part > 0:
            binary_integer = str(integer_part % 2) + binary_integer
            integer_part //= 2

        binary_fractional = ""
        max_precision = 10  # You can adjust the precision as needed
        precision = 0

        while fractional_part > 0 and precision < max_precision:
            fractional_part *= 2
            bit = int(fractional_part)
            binary_fractional += str(bit)
            fractional_part -= bit
            precision += 1

        binary_representation = binary_integer
        if binary_fractional:
            binary_representation += "." + binary_fractional

        print(f'The binary form of {decimal} is: {binary_representation}')
```

# Question 4. Ramanujan-Hardy-Taxi

#### Method 1

```python
def find_interesting_number(N):
    found = False

    for p in range(1, N):
        for q in range(p, N):
            for r in range(p + 1, N):
                for s in range(r, N):
                    if p != q and r != s and p ** 3 + q ** 3 == r ** 3 + s ** 3 and p ** 3 + q ** 3 < N:
                        found = True
                        print(f"Interesting number found: {p}^3 + {q}^3 = {r}^3 + {s}^3")

    if not found:
        print("No interesting number found for the given limit N.")

try:
    N = int(input("Enter a positive integer N: "))
    if N <= 0:
        print("Please enter a positive integer.")
    else:
        find_interesting_number(N)
except ValueError:
    print("Invalid input. Please enter a positive integer.")
```

#### Method 2: much faster

```python
def find_interesting_number(N):
    found = False
    cube_dict = {}

    for p in range(1, N):
        for q in range(p, N):
            cube_sum = p ** 3 + q ** 3
            if cube_sum < N:
                if cube_sum not in cube_dict:
                    cube_dict[cube_sum] = [(p, q)]
                else:
                    cube_dict[cube_sum].append((p, q))

    for r in range(1, N):
        for s in range(r, N):
            cube_sum = r ** 3 + s ** 3
            if cube_sum in cube_dict:
                for pair in cube_dict[cube_sum]:
                    p, q = pair
                    if p != r and p != s and q != r and q != s:
                        found = True
                        print(f"Interesting number found: {p}^3 + {q}^3 = {r}^3 + {s}^3:",r ** 3 + s ** 3)

    if not found:
        print("No interesting number found for the given limit N.")

try:
    N = int(input("Enter a positive integer N: "))
    if N <= 0:
        print("Please enter a positive integer.")
    else:
        find_interesting_number(N)
except ValueError:
    print("Invalid input. Please enter a positive integer.")
```
