## 5: Binary to Decimal

```python
def binary_to_decimal(binary_str):
    decimal_num = 0
    binary_str = binary_str[::-1]  # Reverse the binary string for easy iteration

    for i in range(len(binary_str)):
        if binary_str[i] == '1':
            decimal_num += 2 ** i

    return decimal_num


binary_input = input("Enter a binary number: ")
try:
    decimal_output = binary_to_decimal(binary_input)
    print(f"The decimal equivalent of {binary_input} is: {decimal_output}")
except ValueError:
    print("Invalid binary number. Please enter a valid binary number.")
```

## 6: Integer to Binary

```python
#Positive and Negative
def decimal_to_binary(num):
    if num < 0:
        # Calculate the two's complement of a negative number
        num = 2**32 + num  # Assuming 32-bit integers

    binary = ""
    while num > 0:
        binary = str(num % 2) + binary
        num //= 2

    return binary

try:
    num = int(input('Enter an integer: '))
    binary = decimal_to_binary(num)
    print(f'The binary form of {num} is: {binary}')
except ValueError as e:
    print(e)
```

## 7: Decimal to Binary

```python
try:
    decimal = float(input('Enter a decimal number: '))
except ValueError as e:
    print('Invalid input:', e)
else:
    if decimal < 0:
        print("Please enter a non-negative decimal number.")
    else:
        integer_part = int(decimal)
        fractional_part = decimal - integer_part

        binary_integer = ""
        while integer_part > 0:
            binary_integer = str(integer_part % 2) + binary_integer
            integer_part //= 2

        binary_fractional = ""
        max_precision = 10
        precision = 0

        while fractional_part > 0 and precision < max_precision:
            fractional_part *= 2
            bit = int(fractional_part)
            binary_fractional += str(bit)
            fractional_part -= bit
            precision += 1

        binary_representation = binary_integer
        if binary_fractional:
            binary_representation += "." + binary_fractional

        print(f'The binary form of {decimal} is: {binary_representation}')
```

## 9: Quadratic Equation

```python
def quadratic_roots(a, b, c):
    # Calculate the discriminant
    disc = b**2 - 4*a*c

    # Calculate the roots
    if disc >= 0:
        xp = (-b + np.sqrt(disc)) / (2*a)
        xm = (-b - np.sqrt(disc)) / (2*a)
        return xp, xm

# Generate values for 'a'
a_values = np.linspace(1e-2, 25e-2, 100)

# Initialize lists to store roots
positive_roots = []
negative_roots = []

# Calculate roots for each 'a'
for a in a_values:
    b = 1  # Example value for 'b'
    c = 1  # Example value for 'c'

    xp, xm = quadratic_roots(a, b, c)

    positive_roots.append(xp)
    negative_roots.append(xm)

# Plot the positive and negative roots as a function of 'a'
plt.plot(a_values, positive_roots, label='Positive Root')
plt.plot(a_values, negative_roots, label='Negative Root')
plt.xlabel('a')
plt.ylabel('Roots')
plt.title('Quadratic Equation Roots as a Function of a')
plt.legend()
plt.grid(True)
plt.show()
```

## 10 & 11: Fibonacci Series, Golden Mean, evaluate $\phi^n$

```python
# Function to calculate the nth Fibonacci number
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b

# Generate Fibonacci sequence and calculate the golden mean
n_values = np.arange(2, 21)
fibonacci_values = [fibonacci(n) for n in n_values]
phi_values = [fibonacci(n) / fibonacci(n - 1) for n in n_values]

# Plot Fibonacci sequence
plt.plot(n_values, fibonacci_values, label='Fibonacci Sequence', color = 'navy', marker='o')
plt.xlabel('n')
plt.ylabel('Fibonacci Number')
plt.title('Fibonacci Sequence')
plt.legend()
plt.grid(True)
plt.show()

# Plot golden mean (phi)
# Plot the stability of the golden mean as n increases
plt.plot(n_values, phi_values, marker='o', linestyle='-', color='navy')
plt.axhline(y=(1 + np.sqrt(5)) / 2, color='r', linestyle='--', label='True Golden Mean')
plt.xlabel('n')
plt.ylabel('Golden Mean ($\phi$)')
plt.title('Stability of the Golden Mean ($\phi$) as n Increases')
plt.legend()
plt.grid(True)
plt.show()
```

## 12: Square Root

```python
def square_root(x):
    if x < 0:
        raise ValueError("Cannot calculate the square root of a negative number.")
    
    guess = x / 2.0
    epsilon = 1e-10  # A very small number for precision
    
    while abs(guess * guess - x) > epsilon:
        guess = (guess + x / guess) / 2.0
    
    return guess

# Example usage
number = float(input('Enter a number: '))
result = square_root(number)

print(f"Square root of {number}: {result}")
```

## 13: Root Finding

### Bisection

```python
def f(x):
    return x**2 - 2*np.sin(x)

# Bisection method to find the root within the specified interval [a, b]
def bisection_method(a, b, f, tol=1e-6, max_iterations=1000):
    if f(a) * f(b) >= 0:
        raise ValueError("Bisection method requires the function to change sign over the interval.")

    error_estimate = np.inf
    x_root = None
    iterations = 0

    while error_estimate > tol and iterations < max_iterations:
        c = (a + b) / 2
        error_estimate = np.abs(b - a) / 2
        if f(c) == 0:
            x_root = c
            break
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
        iterations += 1

    if x_root is None:
        x_root = (a + b) / 2

    return x_root, error_estimate, iterations

# Given interval [a, b] where the function changes sign
a = 1.0
b = 3.0 

# Finding the root using the bisection method
root_bisection, error_estimate_bisection, iterations_bisection = bisection_method(a, b, f)

# Printing the root and error estimate using the bisection method
print(f"Root using Bisection Method: {root_bisection:.6f}")
print(f"Error Estimate using Bisection Method: {error_estimate_bisection:.6e} ")
print("Number of Iterations using Bisection Method:", iterations_bisection)
```

### Newton-Raphson

```python
# Defining the function and its derivative
def f(x):
    return x**3 - 4*x*np.sin(x) + 2*np.sin(x)

def dfdx(x):
    return 3*x**2 - 4*x*np.cos(x) - 4*np.sin(x) + 2*np.cos(x)

# Newton-Raphson method with error estimation
def newton_raphson(f, dfdx, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x) / dfdx(x)
        absolute_error = abs(x_new - x)
        
        if absolute_error < tol:
            return x_new, absolute_error, i + 1  # Root found
        
        x = x_new

        # Estimate error using the formula
        error_estimate = abs(dfdx(x_new)) * absolute_error**2
        
        if error_estimate < tol:
            return x_new, error_estimate, i + 1  # Root found

    raise ValueError("Newton-Raphson did not converge within the specified number of iterations.")

x0 = 2  # Initial guess
tolerance = 1e-6
max_iterations = 100

root, error, iterations = newton_raphson(f, dfdx, x0, tol=tolerance, max_iter=max_iterations)

print(f"The root of the equation is approximately {root:.6f} after {iterations} iterations.")
print(f"Estimated error using the formula: {error:.6e}")
```

## 14: Complex Variables: Newton's Fractal

```python
# Define the function and its derivative
def f(z):
    return z**3 - 1

def df(z):
    return 3 * z**2

# Newton's method iteration
def newton_iteration(z, nmax, tolerance):
    n = 0
    while abs(f(z)) > tolerance and n < nmax:
        z = z - f(z) / df(z)
        n += 1
    return n

# Define parameters
nmax = 80
tolerance = 1e-6

# Define the grid
x1, x2 = -2, 2
y1, y2 = -2, 2
N = 1000
x = np.linspace(x1, x2, N)
y = np.linspace(y1, y2, N)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Initialize array to store iteration counts
iteration_counts = np.zeros((N, N))

# Iterate over each point in the grid
for i in range(N):
    for j in range(N):
        z = Z[i, j]
        iteration_counts[i, j] = newton_iteration(z, nmax, tolerance)

# Plot the Newton's Fractal
plt.imshow(iteration_counts, extent=(x1, x2, y1, y2), cmap='rainbow', origin='lower')
plt.colorbar(label='Iteration Count')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title("Newton's Fractal")
plt.show()
```

### Trapezoidal

```python
def trap_int(f, a=0, b=1, e=1e-10):
    # Initial step size
    h = (b - a)
    I_prev = 0

    # Iterate until the error is within the tolerance
    while True:
        N = int((b - a) / h)
        I = 0.5 * (f(a) + f(b))  # Initialize the integral

        # Calculate the trapezoidal rule
        for i in range(1, N):
            I += f(a + i * h)

        I *= h

        # Check for convergence based on tolerance
        if abs(I - I_prev) < e:
            return I

        # Update values for the next iteration
        I_prev = I
        h /= 2
        
result = trap_int(f)
print("Approximate integral by Trapezoidal Integration:", result)
```

### Simpsons

```python
def simp_int(f, a=0, b=1, e=1e-10):
    N = 2  # Initial number of intervals (N=2 for Simpson's rule)
    h = (b - a) / N
    I_prev = 0  # Previous estimate of the integral

    while True:
        # Calculate the integral using Simpson's rule
        I = f(a) + f(b)  # Initialize integral

        for i in range(1, N, 2):  # Odd indices
            I += 4 * f(a + i * h)

        for i in range(2, N - 1, 2):  # Even indices (excluding last point)
            I += 2 * f(a + i * h)

        I *= h / 3

        # Check for convergence based on tolerance
        if abs(I - I_prev) < e:
            return I

        # Update values for the next iteration
        I_prev = I
        N *= 2  # Double the number of intervals
        h /= 2  # Halve the step size
        
result = simp_int(f)
print("Approximate integral by Simpson Integration:", result)
```

## 17: Adaptive Integration

```python
def integrate_adaptive(f, a, b, fa, fb, epsilon, max_depth=1000):
    # Midpoint and trapezoidal approximations
    m = (a + b) / 2
    fm = f(m)
    approx_mid = (fa + fb + 2 * fm) * (b - a) / 4
    approx_trap = (fa + fb) * (b - a) / 2

    # Estimate the error
    error = abs(approx_mid - approx_trap) / 3

    # If error is within tolerance or maximum depth is reached, return the approximation
    if error < epsilon or max_depth == 0:
        return approx_mid, 1

    # Recursive call for the left and right subintervals
    left_approx, left_iterations = integrate_adaptive(f, a, m, fa, fm, epsilon/2, max_depth-1)
    right_approx, right_iterations = integrate_adaptive(f, m, b, fm, fb, epsilon/2, max_depth-1)

    return left_approx + right_approx, left_iterations + right_iterations

# Define the integration interval
a = 0
b = 1

# Calculate function values at the endpoints to reduce function evaluations
fa = f(a)
fb = f(b)

# Set tolerance levels
epsilon = 1e-6

# Initialize lists to store the results
integral_results = []
iteration_counts = []

# Perform adaptive integration for each tolerance level
integral, iterations = integrate_adaptive(f, a, b, fa, fb, epsilon)
print(f"Integral approximation for epsilon 1e-6= : {integral}, Iterations: {iterations}")
```

## 22: Area of Circle using Random Number

```python
def euclidean_norm(point):
    return (point[0]**2 + point[1]**2)**0.5

def monte_carlo_circle_area(num_points, radius):
    # Generate random points within the square [-radius, radius] x [-radius, radius]
    random_points = [(2 * radius * np.random.uniform(-0.5, 0.5), 2 * radius * random.uniform(-0.5, 0.5)) for _ in range(num_points)]

    # Check if points fall within the circle
    count_inside_circle = 0

    for point in random_points:
        if euclidean_norm(point) <= radius:
            count_inside_circle += 1

    # Calculate the ratio of points inside the circle to the total points
    ratio_inside_circle = count_inside_circle / num_points

    # Calculate the estimated area of the circle
    estimated_area = ratio_inside_circle * (2 * radius)**2

    # Visualize the points and the circle
    plt.figure(figsize=(6, 6))
    inside_circle_points = [point for point in random_points if euclidean_norm(point) <= radius]
    outside_circle_points = [point for point in random_points if euclidean_norm(point) > radius]

    inside_x, inside_y = zip(*inside_circle_points)
    outside_x, outside_y = zip(*outside_circle_points)

    plt.scatter(inside_x, inside_y, color='blue', marker='.', label='Inside Circle')
    plt.scatter(outside_x, outside_y, color='red', marker='.', label='Outside Circle')
    circle = plt.Circle((0, 0), radius, edgecolor='black', facecolor='none', linewidth=2)
    plt.gca().add_patch(circle)
    plt.title('Monte Carlo Simulation of Circle Area')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()

    return estimated_area

# Parameters
num_points = 10000
radius = 1.0

# Calculate the estimated area of the circle using Monte Carlo simulation
estimated_area = monte_carlo_circle_area(num_points, radius)

# Print the result
print(f"Estimated Area of the Circle with radius 1: {estimated_area}")
```

### Importance Sampling

```python
# Importance Sampling Monte Carlo Method for Gaussian Distribution
def importance_sampling_monte_carlo_gaussian(num_samples=100000):
    integral_estimate = 0
    normalization_factor = np.sqrt(2 * np.pi)  # Normalization factor for p(x)

    for _ in range(num_samples):
        # Generating a sample from the probability density function p(x)
        u = np.random.uniform(0, 1)
        x = np.sqrt(-2 * np.log(u))

        # Evaluating the integrand f(x) = e^(-x^2)
        f_x = np.exp(-x**2)

        # Evaluating the probability density function p(x) = 1 / sqrt(2*pi)
        p_x = 1 / normalization_factor

        # Calculating the weight for importance sampling
        weight = f_x / p_x

        # Updating the integral estimate
        integral_estimate += weight

    # Calculating the final estimate of the integral
    estimate = integral_estimate / num_samples
    return estimate

# Performing the simulation for Gaussian Distribution
num_samples = 10000
gaussian_importance_estimate = importance_sampling_monte_carlo_gaussian(num_samples)
print("Gaussian Distribution:")
print("Importance Sampling Monte Carlo Estimate:", gaussian_importance_estimate)

# Plotting
x_values = np.linspace(-3, 3, 10000)
y_values = np.exp(-x_values**2) / np.sqrt(2 * np.pi)

plt.hist(np.sqrt(-2 * np.log(np.random.uniform(0, 1, num_samples))), bins=30, density=True, label='Generated Samples',color='navy')
plt.title('Importance Sampling for Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

## 24: Least Square Fit

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Data
m = 2
c = 3
x = np.arange(0, 50)
y = m * x + c
y_noisy = y + np.random.uniform(0, 5, len(y))  # No bias given, just adding random noise

# Least Squares Fit
def least_squares_fit(x_values, y_values, degree):
    num_points = len(x_values)
    sum_x = [np.sum(x_values ** i) for i in range(2 * degree + 1)]
    sum_y = [np.sum(y_values * (x_values ** i)) for i in range(degree + 1)]

    A = np.zeros((degree + 1, degree + 1))
    for i in range(degree + 1):
        A[i, :] = sum_x[i:i + degree + 1]

    coefficients = np.linalg.solve(A, sum_y)

    return coefficients

degree_of_fit = 1  # Degree of the polynomial fit
coefficients = least_squares_fit(x, y_noisy, degree_of_fit)

# Fitted Line
fitted_line = np.poly1d(coefficients[::-1])

# Plotting
plt.scatter(x, y_noisy, label='Noisy Data',color='navy')
plt.plot(x, fitted_line(x), 'r', label='Fitted Line')
plt.plot(x, m * x + c, 'g', label='True Line')
plt.legend()
plt.title('Data and Fitted Line')
plt.show()
```

## 25: Interpolation and Extrapolation

```python
def lagrange_interpolation(x_values, y_values, x):
    result = 0
    for i in range(len(y_values)):
        term = y_values[i]
        for j in range(len(x_values)):
            if j != i:
                term = term * (x - x_values[j]) / (x_values[i] - x_values[j])
        result += term
    return result

x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 1, 3, 6, 8])

# Interpolation
x_interpolate = 2.5
y_interpolate = lagrange_interpolation(x_data, y_data, x_interpolate)
print(f'Interpolated value at {x_interpolate}: {y_interpolate}')

# Extrapolation
x_extrapolate = 6
y_extrapolate = lagrange_interpolation(x_data, y_data, x_extrapolate)
print(f'Extrapolated value at {x_extrapolate}: {y_extrapolate}')

# Plotting
x_values = np.linspace(min(x_data), max(x_data), 100)
y_values = [lagrange_interpolation(x_data, y_data, x) for x in x_values]

plt.scatter(x_data, y_data, label='Data Points', color='green')
plt.plot(x_values, y_values, label='Interpolation Curve', color='navy')
plt.title('Lagrange Polynomial Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 26: Differential Equations

$\frac{d^2x}{dt^2} = -kx$

where \( $k$ \) is a constant. We can convert this second-order differential equation into a system of two first-order differential equations:

$\frac{dx}{dt} = v $

$\frac{dv}{dt} = -kx$

Implementing the Euler method, RK-2, and RK-4 to solve this system of differential equations and plot \( $x$ \) vs \( $t$ \) for each case:

```python
# Function to calculate the derivatives 
def derivatives(t, u, k):
    x, v = u
    dxdt = v
    dvdt = -k * x
    return [dxdt, dvdt]

# Euler method
def euler_method(t0, u0, k, h, num_steps):
    t_values = [t0]
    u_values = [u0]

    for _ in range(num_steps):
        u0 = [u0[i] + h * derivatives(t0, u0, k)[i] for i in range(2)]
        t0 += h

        t_values.append(t0)
        u_values.append(u0)

    return t_values, u_values

# Runge-Kutta 2nd order method
def runge_kutta_2(t0, u0, k, h, num_steps):
    t_values = [t0]
    u_values = [u0]

    for _ in range(num_steps):
        k1 = [h * val for val in derivatives(t0, u0, k)]
        k2 = [h * val for val in derivatives(t0 + 0.5 * h, [u0[i] + 0.5 * k1[i] for i in range(2)], k)]

        u0 = [u0[i] + k2[i] for i in range(2)]
        t0 += h

        t_values.append(t0)
        u_values.append(u0)

    return t_values, u_values

# Runge-Kutta 4th order method
def runge_kutta_4(t0, u0, k, h, num_steps):
    t_values = [t0]
    u_values = [u0]

    for _ in range(num_steps):
        k1 = [h * val for val in derivatives(t0, u0, k)]
        k2 = [h * val for val in derivatives(t0 + 0.5 * h, [u0[i] + 0.5 * k1[i] for i in range(2)], k)]
        k3 = [h * val for val in derivatives(t0 + 0.5 * h, [u0[i] + 0.5 * k2[i] for i in range(2)], k)]
        k4 = [h * val for val in derivatives(t0 + h, [u0[i] + k3[i] for i in range(2)], k)]

        u0 = [u0[i] + (k1[i] + 2 * k2[i] + 2 * k3[i] + k4[i]) / 6 for i in range(2)]
        t0 += h

        t_values.append(t0)
        u_values.append(u0)

    return t_values, u_values

# Initial conditions
t0 = 0
u0 = [1, 0]  # Initial position and velocity
k = 1  # Constant in the differential equation
h = 0.1  # Step size
num_steps = 100

# Solve using Euler method
t_euler, u_euler = euler_method(t0, u0, k, h, num_steps)

# Solve using Runge-Kutta 2nd order method
t_rk2, u_rk2 = runge_kutta_2(t0, u0, k, h, num_steps)

# Solve using Runge-Kutta 4th order method
t_rk4, u_rk4 = runge_kutta_4(t0, u0, k, h, num_steps)

# Extract positions for plotting
x_euler = [u[0] for u in u_euler]
x_rk2 = [u[0] for u in u_rk2]
x_rk4 = [u[0] for u in u_rk4]

# Plotting
plt.plot(t_euler, x_euler, label='Euler Method', color='navy')
plt.plot(t_rk2, x_rk2, label='Runge-Kutta 2nd Order', color='green')
plt.plot(t_rk4, x_rk4, label='Runge-Kutta 4th Order', color='orange', linestyle='--')
plt.xlabel('Time ($t$)')
plt.ylabel('Position ($x$)')
plt.legend()
plt.show()
```

## 27: Ising Model

```python
def initialize_spins(L):
    """Initialize a random configuration of spins in a 2D lattice."""
    spins = np.random.choice([-1, 1], size=(L, L))
    return spins

def calculate_energy(spins):
    """Calculate the energy of the spin configuration in a 2D lattice."""
    energy = -np.sum(spins * (np.roll(spins, 1, axis=0) + np.roll(spins, 1, axis=1)))
    return energy

def metropolis_update(spins, beta):
    """Perform a Metropolis update on a randomly chosen spin in a 2D lattice."""
    L = spins.shape[0]
    i, j = np.random.randint(0, L, size=2)

    # Calculate energy change if the spin is flipped
    delta_E = 2 * spins[i, j] * (
        spins[(i + 1) % L, j] + spins[(i - 1) % L, j] +
        spins[i, (j + 1) % L] + spins[i, (j - 1) % L]
    )

    # Metropolis condition
    if delta_E < 0 or np.random.rand() < np.exp(-beta * delta_E):
        spins[i, j] *= -1  # Flip the spin if conditions are met

def ising_model(L, num_steps, beta):
    """Simulate the 2D Ising model."""
    spins = initialize_spins(L)
    energies = []

    for _ in range(num_steps):
        metropolis_update(spins, beta)
        energy = calculate_energy(spins)
        energies.append(energy)

    return spins, energies

def plot_ising(spins):
    """Plot the spin configuration in a 2D lattice."""
    plt.imshow(spins, cmap='binary', vmin=-1, vmax=1)
    plt.xticks([]), plt.yticks([])
    plt.title('2D Ising Model Spin Configuration')
    plt.show()

def plot_energy(energies):
    """Plot the energy evolution during the simulation."""
    plt.plot(energies, color='k')
    plt.xlabel('Step')
    plt.ylabel('Energy')
    plt.title('Energy Evolution in the 2D Ising Model')
    plt.show()

# Parameters
L = 100  # Size of the 2D lattice
num_steps = 1000
beta = 1.0  # Inverse temperature

# Simulate and plot
spins, energies = ising_model(L, num_steps, beta)
plot_ising(spins)
plot_energy(energies)
```

## 28: SIR Model

```python
# Initial population and parameters
population = 1000
infected = 1
recovered = 0
susceptible = population - infected - recovered
beta = 0.4  # Infection rate
gamma = 0.05  # Recovery rate

# Lists to store data for plotting
susceptible_list = [susceptible]
infected_list = [infected]
recovered_list = [recovered]

# Simulation for 200 days
for day in range(200):
    new_infected = beta * infected * susceptible / population
    new_recovered = gamma * infected

    susceptible -= new_infected
    infected += new_infected - new_recovered
    recovered += new_recovered

    susceptible_list.append(susceptible)
    infected_list.append(infected)
    recovered_list.append(recovered)

# Plotting the SIR curves
plt.figure(figsize=(8, 6))
plt.plot(susceptible_list, label='Susceptible', color='navy')
plt.plot(infected_list, label='Infected', color='red')
plt.plot(recovered_list, label='Recovered',color='green')
plt.title('Basic SIR Model')
plt.xlabel('Days')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```
